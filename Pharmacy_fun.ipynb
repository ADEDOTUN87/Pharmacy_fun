# python script that defines a function called womens_supplement

```python
# Define a function women's_supplement
def womens_supplement():
    """
    Returns a list of supplements for women's health.
    """
    # Define the supplements
    supplements = {
        "Multivitamin": "Daravit Perfect 3",
        "Skin & Hair": "Evening Primrose",
        "Gut Health": "Probiotic Gummies"
    }
# Create a variable perfect_combo containing the supplements
    perfect_combo = list(supplements.values())
 # Return the perfect combo
    return perfect_combo

```

```python
 
# Call the function and store the result in a variable
perfect_combo = womens_supplement()

```

```python
# Print the perfect combo
print("Perfect Combo:")
for i, supplement in enumerate(perfect_combo, start=1):
    print(f"{i}. {supplement}")
```

```python
# conditional statements
def womens_supplement(supplement):
    womens_supplement = ["Daravit Perfect 3", "Evening Primrose", "Probiotic Gummies"]
    if supplement in womens_supplement:
        return f"{supplement} is a women's supplement"
    else:
        return "Not for women"
```

```python
# Test the function
print(women_supplement("Daravit Perfect 3"))  # Daravit Perfect 3 is a women's supplement
print(women_supplement("Reload Men"))  # Not for women
print(women_supplement("Centrum Men"))  # Not for women
print(women_supplement("Daravit Man"))  # Not for women
print(women_supplement("Revive"))  # Not for women
print(women_supplement("Manix Capsule"))  # Not for women
```

```python
# Function to check if a supplement is for women

def womens_supplement():
    return ["Daravit Perfect 3", "Evening Primrose Oil", "Probiotic Gummies"]
```

```python
def is_for_women(supplement_name):
    perfect_combo = womens_supplement()
    if supplement_name in perfect_combo:
        return f"{supplement_name} is recommended for women."
    else:
        return "Not for women"

# Example calls
print(is_for_women("Daravit Perfect 3"))     # Output: recommended
print(is_for_women("Reload Men"))            # Output: Not for women
print(is_for_women("Manix Capsule"))         # Output: Not for women
```
